```python
import pandas as pd 
import numpy as np
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
from sklearn import datasets
```

```python
data1 = datasets.load_iris()
```

```python
df = pd.DataFrame(data1.data,columns = data1.feature_names)
df
```

```python
X = df.iloc[:,[0,1,2,3]].values
X
```

```python
wss=[]
K=range(1,12)
for k in K:
    km = KMeans(n_clusters=k,init='k-means++', )
    km.fit(X)
    wss_iner = km.inertia_
    wss.append(wss_iner)
    
    
```

```python
plt.plot(K,wss,marker='+',color='red')
plt.xlabel("Clusters")
plt.ylabel("WSS")
plt.grid()
plt.show
```

```python
kmeans = KMeans(n_clusters=3, random_state=0)

y_predicted = kmeans.fit_predict(X)
y_predicted
```

```python
df['Cluster'] = y_predicted
df.head(10)
```

```python
df1 = df[df['Cluster']==0]
df2 = df[df['Cluster']==1]
df3 = df[df['Cluster']==2]
```

```python
plt.scatter(df1['petal length (cm)'],df1['petal width (cm)'],color="red",label='Iris- Setosa',marker='+')
plt.scatter(df2['petal length (cm)'],df2['petal width (cm)'],color="green",label='Iris- VesiColour',marker='+')
plt.scatter(df3['petal length (cm)'],df3['petal width (cm)'],color="blue",label='Iris- Verginicia',marker='+')
plt.scatter(kmeans.cluster_centers_[:,2] , kmeans.cluster_centers_[:,3], color = 'purple',marker='P')
plt.title("Cluster Visualization")
plt.legend()
plt.show()
```
